##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.

```python
from collections import defaultdict
```

```python
res_dict = defaultdict(list)
```

```python
format_df = df[["Mjob","Medu"]]
format_df
```

```python
for idx,row in format_df.iterrows():
  res_dict[row[0]].append(row[1])
```

```python
for n,m in res_dict.items():
  print(f"Категория - ({n}) min = {min(m)} mean = {sum(m)/len(m)} ")
```
