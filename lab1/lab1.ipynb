##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.

```python
data = [
    {"category":"A","num":10},
    {"category":"B","num":20},
    {"category":"A","num":12},
    {"category":"B","num":3},
    {"category":"A","num":40},
]

sums = {}
counts = {}
median = {}

for x in data:
  category = x["category"]
  num = x["num"]

  if category in sums:
    sums[category] += num
  else:
    sums[category] = num

  if category in counts:
    counts[category] += 1
  else:
    counts[category] = 1

  if category in median:
    counts[category] += 1
  else:
    counts[category] = 1

averages = {}
for category in sums:
    averages[category] = sums[category] / counts[category]

# Выводим результаты
for category, average in averages.items():
    print(f'Категория: {category}, Среднее значение: {average}')
```
